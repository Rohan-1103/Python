# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\Projects\python\Ice Cream.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('_classic_test_patch')
```

```python
df.plot()
```

```python
df.plot(kind = 'bar')
```

```python
 df.plot(kind = 'line', subplots = True)
```

```python
 df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores', figsize = (10, 5))
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'orange')
```

```python
df.plot.hist(bins = 30)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10, 6))
```
