```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = ['description', 'neighborhood_overview', 'medium_url', 'picture_url', 'xl_picture_url', 'host_response_time', 'host_response_rate', 'host_acceptance_rate', 'host_neighbourhood', 'review_scores_accuracy'])
```

```python
df['id'].apply(lambda x: str(x))
df['id'] = df['id'].apply(lambda x: x[0:1] + '-' + x[1:3] + '-' + x[3:7])
```

```python
df
```
