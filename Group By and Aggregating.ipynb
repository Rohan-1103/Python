# Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\vgSales.csv")
df
```

```python
group_by_frame = df.groupby('Platform')
```

```python
group_by_frame.max('NA_Sales')
```

```python
group_by_frame.min('NA_Sales')
```

```python
group_by_frame.mean('NA_Sales')
```

```python
df.groupby('Platform').agg({'NA_Sales': ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby('Platform').describe()
```
